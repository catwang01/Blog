# Scipy sparse matrix

```python
import numpy as np
from scipy import sparse

x = np.random.randn(30, 20)
x_csr = sparse.csr_matrix(x)
```

```python
x_csr.toarray() # 转换回 array
```

csr 格式的稀疏矩阵可以切片，返回的仍然是稀疏矩阵

```python
x_csr[:10, 2:4]
```

而相比之下， coo 格式的稀疏矩阵不能切片

```python
x_coo = sparse.coo_matrix(x)

x_coo[:10, 2:4]
```

coo 格式可以转化为 csr 格式

```python
x_coo.tocsr()
```
