# Matplotlib latex




在 matplotlib 中可以直接使用 latex，只需要在两侧添加 `$` 即可。

```python
import numpy as np
from matplotlib import pyplot as plt

x=np.arange(0.01,1,0.01)
y=0.5*np.log((1-x)/x)

plt.plot(x,y,label=r'$\alpha =\frac{1}{2}\ln(\frac{1-\varepsilon}{\varepsilon })$')

plt.legend()
plt.xlabel(r'$\varepsilon$', fontsize=20)
plt.ylabel(r'$\alpha$', fontsize=20)   

plt.show()
```
